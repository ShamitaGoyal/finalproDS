## 2. Clean Data
- checking for NaNs
- simplifying the column labels

```python
# checking for any NaNs
df.isna().sum()
```

```python
# changing the column labels so they are more readable
# making the column labels into lower case
df.columns = df.columns.str.lower()
df.columns
```

```python
# changing the labels of the columns that are too wordy
labels = {
    "state/ut":"state",
    "district":"dist",
    "attempt to murder":"attem_murder",
    "culpable homicide not amounting to murder":"cul_homicide", 
    "kidnapping & abduction": "kidnap_abd",
    "kidnapping and abduction of women and girls": "kidnap_abd_female",
    "kidnapping and abduction of others": "kidnap_abd_other",
    "preparation and assembly for dacoity": "prep_dacoity",
    "criminal breach of trust": "CBT",
    "hurt/grevious hurt":"hurt",
    "assault on women with intent to outrage her modesty":"assault_wom",
    "insult to modesty of women":"insult_wom",
    "cruelty by husband or his relatives":"cruel_hus",
    "importation of girls from foreign countries": "imp_girls",
    "causing death by negligence": "death_neg",
    "other ipc crimes": "other_ipc",
    "total ipc crimes":"total_ipc"
}

df = df.rename(columns=labels)

# changing the state and dist column values into title case
df.state = df.state.str.title()
df.dist = df.dist.str.title()
```
